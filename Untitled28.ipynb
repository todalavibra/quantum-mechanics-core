```python
import numpy as np
import matplotlib.pyplot as plt

# --- SETTING THE STAGE ---
# We are simulating a wave passing through two small slits.
# Imagine dropping two pebbles into a pond side-by-side; the ripples will overlap.

def double_slit_simulation():
    # 1. Define the space (The screen where we look for particles)
    # 1000 points along the screen
    x = np.linspace(-10, 10, 1000)

    # 2. Define the Physics Parameters
    wavelength = 1.0       # The "color" or energy of the wave
    slit_distance = 3.0    # How far apart the two slits are
    k = 2 * np.pi / wavelength  # The "wave number" (spatial frequency)

    # 3. Calculate the Wave Functions from each slit
    # r1 is distance from slit 1 to point x on screen
    # r2 is distance from slit 2 to point x on screen
    # Using small angle approximation for simplicity (valid for far field)
    r1 = np.sqrt(x**2 + (slit_distance/2)**2)
    r2 = np.sqrt(x**2 + (-slit_distance/2)**2)

    # The wave function (Psi) is a complex number: exp(i * k * r)
    psi_1 = np.exp(1j * k * r1)  # Wave from top slit
    psi_2 = np.exp(1j * k * r2)  # Wave from bottom slit

    # 4. Superposition Principle: Add the waves together!
    psi_total = psi_1 + psi_2

    # 5. Probability Density (The "Interference Pattern")
    # In quantum mechanics, Probability = |Psi|^2
    probability = np.abs(psi_total)**2

    # --- VISUALIZATION ---
    plt.figure(figsize=(12, 6))
    plt.style.use('dark_background') # For that cool physics lab look

    # Plotting the Probability Density (The Pattern you see)
    plt.plot(x, probability, color='#00ffcc', linewidth=2, label='Electron Probability')
    plt.fill_between(x, probability, color='#00ffcc', alpha=0.3)

    # Add labels and "Slit" markers for context
    plt.title('Double Slit Experiment: Interference Pattern', fontsize=16)
    plt.xlabel('Position on Detector Screen', fontsize=12)
    plt.ylabel('Probability of Finding Particle', fontsize=12)

    # Drawing lines to show where the slits are "located" relative to the pattern
    plt.axvline(-slit_distance/2, color='white', linestyle='--', alpha=0.5, label='Slit 1 Position')
    plt.axvline(slit_distance/2, color='white', linestyle='--', alpha=0.5, label='Slit 2 Position')

    plt.legend(loc='upper right')
    plt.grid(True, alpha=0.2)
    plt.show()

# Run the simulation
if __name__ == "__main__":
    double_slit_simulation()

import numpy as np
import matplotlib.pyplot as plt

def particle_in_a_box(n_level):
    """
    Visualizes the probability of finding a particle in a 1D box.
    n_level: The energy level (1, 2, 3...)
    """
    # 1. Define the Box (0 to L)
    L = 1.0  # Width of the box
    x = np.linspace(0, L, 100)

    # 2. The Wave Function Formula for a Box
    # Psi_n(x) = sqrt(2/L) * sin(n * pi * x / L)
    psi = np.sqrt(2/L) * np.sin(n_level * np.pi * x / L)

    # 3. Probability Density (|Psi|^2)
    prob = psi**2

    # 4. Plotting
    plt.figure(figsize=(10, 4))
    plt.style.use('dark_background')

    # Plot the probability
    plt.plot(x, prob, color='lime', linewidth=2, label=f'Probability (Energy Level n={n_level})')
    plt.fill_between(x, prob, color='lime', alpha=0.3)

    # Draw the walls of the box
    plt.axvline(0, color='red', linewidth=4, label='Wall')
    plt.axvline(L, color='red', linewidth=4)

    plt.title(f'Particle in a Box: Energy Level {n_level}', fontsize=14)
    plt.xlabel('Position in Box', fontsize=12)
    plt.ylabel('Probability', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.2)
    plt.show()

# --- INSTRUCTIONS ---
# Try changing the number inside the parentheses below!
# 1 = Ground State (Lowest Energy)
# 2 = First Excited State
# 3 = Second Excited State
particle_in_a_box(3)

def visualize_tunneling():
    # 1. Define the Space
    x = np.linspace(-2, 4, 1000)

    # 2. Define the Barrier (The Wall)
    # The barrier exists from x=0 to x=1
    barrier_width = 1.0
    barrier_height = 10.0  # V_0

    # 3. Create the Potential Energy Graph (V)
    # It is 0 everywhere, except 10 inside the barrier
    V = np.zeros_like(x)
    V[(x >= 0) & (x <= barrier_width)] = barrier_height

    # 4. Define the Particle's Energy (E)
    # CRITICAL: Energy (5) is LESS than Barrier Height (10).
    # Classically, it should bounce back.
    E = 5.0

    # 5. Calculate Wave Function (Psi)
    psi = np.zeros_like(x)

    # Region 1: Before the wall (Incoming Wave)
    # It's oscillating (sin/cos)
    k1 = np.sqrt(2 * E) # Wave number outside
    psi[x < 0] = np.cos(k1 * x[x < 0])

    # Region 2: Inside the wall (Tunneling)
    # It decays exponentially! (The "Ghost" effect)
    # kappa is the decay rate
    kappa = np.sqrt(2 * (barrier_height - E))
    # We match the height at x=0 (which is 1.0 from cos(0))
    psi[(x >= 0) & (x <= barrier_width)] = np.exp(-kappa * x[(x >= 0) & (x <= barrier_width)])

    # Region 3: After the wall (Transmitted Wave)
    # It oscillates again, but with much lower amplitude
    amplitude_after = np.exp(-kappa * barrier_width)
    psi[x > barrier_width] = amplitude_after * np.cos(k1 * (x[x > barrier_width] - barrier_width))

    # --- PLOTTING ---
    plt.figure(figsize=(12, 6))
    plt.style.use('dark_background')

    # Plot the Potential Barrier (The Wall)
    plt.fill_between(x, 0, V, color='red', alpha=0.3, label='Potential Barrier (Wall)')
    plt.plot(x, V, color='red', linestyle='--')

    # Plot the Wave Function
    plt.plot(x, psi + E, color='cyan', linewidth=2, label='Wave Function ($\\psi$)')
    # Note: We added E to psi just to lift the graph up so it overlays the barrier nicely

    # Plot the Energy Level
    plt.axhline(E, color='yellow', linestyle=':', label='Particle Energy')

    plt.title(f'Quantum Tunneling: Energy ({E}) < Barrier ({barrier_height})', fontsize=14)
    plt.xlabel('Position', fontsize=12)
    plt.yticks([]) # Hide y-axis numbers for clarity
    plt.text(0.1, E - 1.5, 'Exponential Decay', color='white', rotation=-45)
    plt.text(2.0, E + 0.5, 'Tunneling Success!', color='cyan')

    plt.legend(loc='upper right')
    plt.show()

if __name__ == "__main__":
    visualize_tunneling()


def plot_bloch_sphere(theta, phi):
    """
    Visualizes a Qubit state on the Bloch Sphere.
    theta: Angle down from the North Pole (0 to pi) - Controls 0 vs 1 mix
    phi: Angle around the equator (0 to 2pi) - Controls the 'phase'
    """
    # 1. Create the Sphere (Wireframe)
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    plt.style.use('dark_background')

    # Draw sphere grid
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = np.outer(np.cos(u), np.sin(v))
    y = np.outer(np.sin(u), np.sin(v))
    z = np.outer(np.ones(np.size(u)), np.cos(v))

    # Plot the wireframe surface
    ax.plot_wireframe(x, y, z, color='gray', alpha=0.1)
    ax.plot_surface(x, y, z, color='cyan', alpha=0.05) # Faint glow

    # 2. Calculate the Arrow (The Qubit State vector)
    # Converting spherical coordinates to Cartesian (x,y,z)
    state_x = np.sin(theta) * np.cos(phi)
    state_y = np.sin(theta) * np.sin(phi)
    state_z = np.cos(theta)

    # 3. Draw the Arrow
    ax.quiver(0, 0, 0, state_x, state_y, state_z,
              color='lime', length=1.0, linewidth=3, arrow_length_ratio=0.1)

    # 4. Add Labels and Poles
    ax.text(0, 0, 1.2, '|0 (North Pole)', color='white', fontsize=12, ha='center')
    ax.text(0, 0, -1.2, '|1 (South Pole)', color='white', fontsize=12, ha='center')
    ax.text(0, 0, 0, 'Center', color='black')

    # Add the "Superposition" label if it's not strictly 0 or 1
    if 0.1 < theta < 3.0:
        ax.text(state_x*1.2, state_y*1.2, state_z*1.2,
                f'State |\n(Superposition)', color='lime', fontsize=10)

    # Formatting
    ax.set_title(f'Qubit Visualization\nTheta={theta:.2f}, Phi={phi:.2f}', fontsize=14)
    ax.set_axis_off() # Hide the ugly box axes
    plt.show()

# --- INSTRUCTIONS ---
# Try changing these angles!
# theta = 0      -> Pure |0> State (North Pole)
# theta = 3.14   -> Pure |1> State (South Pole)
# theta = 1.57   -> 50/50 Superposition (Equator)
plot_bloch_sphere(theta=0, phi=0)
```
